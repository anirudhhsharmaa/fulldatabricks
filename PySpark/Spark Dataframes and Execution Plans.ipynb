**Structured API -- Dataframes**

- _DataFrame is the most common Structure API, represented like a table._
- _The table is represented in form of Rows and Columns._
- _DataFrame has schema, which is the metadata for the Columns._
- _Data in DataFrames are in Partitions._
- _DataFrames are immutable._

**Structured API Execution Plan**

 _Types of Execution Plan_
-  _Logical Planning_
-  _Physical Planning_

**Logical Planning**

_Code -> Unresolved Logical Plan -> Resolved Logical Plan (**verified through Catalog**) -> Catalyst Optimizer -> Optimized Logical Plan_

**Physical Planning**

_Once the Optimized Logical Plan is ready -> Spark generates multiple Physical Plan (**based on Cluster and Physical Configurations**)_
_ ->These_plans are run against a Cost Model (**cost is generated for each of the Physical Plan**) -> Best Physical Plan based on the Costing_
