**Ques. What is Spark?**

_Ans. Spark is an Open Source Computing Engine with set of libraries for parallel data processing on Computing Cluster._

_It supports widely used programming languages such as:_
- A. Scala
- B. Python
- C. Java
- D. R

_It process data In Memory (Ram) which makes it 100 times faster than traditional Hadoop Map Reduce._

**Ques. List down the Spark Components**

_Ans_. 

- Low Level API - RDD and Distributed Variables
- Structured API - Dataframes, Datasets and SQL
- Libraries and Ecosystems, Structured Streaming and Advanced Analytics
